```python

df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
df['Year'] = df['Release Date'].dt.year
```

```python
top_movies = df.loc[df.groupby('Year')['World Wide Sales (in $)'].idxmax()]
```

```python
print(top_movies[['Year', 'Title', 'World Wide Sales (in $)']])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(top_movies['Year'], top_movies['World Wide Sales (in $)'], color='blue')
plt.xlabel('Año')
plt.ylabel('Ventas mundiales (en $)')
plt.title('Comportamiento ventas por año')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_expanded = df.explode('Genre')
```

```python
top_genre = df_expanded.groupby('Genre')['World Wide Sales (in $)'].sum().idxmax()
```

```python
sales_top_genre = df_expanded.groupby('Genre')['World Wide Sales (in $)'].sum().max()
```

```python
print(f"El género que se vendió más fue '{top_genre}' con un total de ventas de {sales_top_genre:.2f} dólares.")
```

```python
sales_by_genre = df_expanded.groupby('Genre')['World Wide Sales (in $)'].sum()
```

```python
num_top_genres = 5
top_genres = sales_by_genre.nlargest(num_top_genres)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_genres.index, top_genres, color='skyblue')
plt.xlabel('Género')
plt.ylabel('Ventas mundiales totales (en $)')
plt.title(f'{num_top_genres} géneros con las mayores ventas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
